```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
from nltk.corpus import stopwords
```

Data Cleansing:

```python
import re
import string

def clean_text(text ):
    text =''.join(word for word in text if word not in string.punctuation)
    text1 = ''.join(w.lower() for w in text if not w.isdigit())
    return text1
```

Tokenization, Lemmatization

```python
# Lemmatization of text data
def lemmatize_text(text):
    wordlist = []
    lemmatizer = WordNetLemmatizer()
    # Sentence tokenization
    sentences = nltk.sent_tokenize(str(text))

    intial_sentences = sentences[0:1]
    final_sentences = sentences[len(sentences)-2: len(sentences)-1]

    # Word tokenization
    for sentence in intial_sentences:
        words = nltk.word_tokenize(sentence)
        for word in words:
            wordlist.append(word)

    for sentence in final_sentences:
        words = nltk.word_tokenize(sentence)
        for word in words:
            wordlist.append(word)

    return wordlist
```

Applying Functions on train data

```python
# Apply the functions on Description column

# Clean
data['text'] = data['text'].apply(lambda x: clean_text(x))
# Lemmatize
data['text'] = data['text'].apply(lambda x: lemmatize_text(x))
```

Using Bert

```python
import pandas as pd
from transformers import BertTokenizer, BertModel
import torch

# Assuming you already have the 'data' DataFrame with a column named 'text'

# Combine text from 'text' column into a single string
data['Combined_Text'] = data['text'].apply(lambda x: ' '.join(map(str, x)))

# Load pre-trained BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Tokenize the combined text
tokenized_texts = data['Combined_Text'].apply(lambda x: tokenizer.encode(x, add_special_tokens=True))

# Pad sequences to a fixed length (assuming a max length of 100 tokens)
max_len = 100
padded_texts = torch.tensor([text[:max_len] + [0]*(max_len-len(text)) for text in tokenized_texts])

# Load pre-trained BERT model
model = BertModel.from_pretrained('bert-base-uncased')

# Set model to evaluation mode
model.eval()

# Generate BERT embeddings
with torch.no_grad():
    outputs = model(padded_texts)
    embeddings = outputs.last_hidden_state[:, 0, :]  # Use the [CLS] token embeddings

# Convert embeddings to a NumPy array
bert_embeddings = embeddings.numpy()

# Now 'bert_embeddings' contains the embeddings for your text data
print(bert_embeddings.shape)  # Output shape: (number_of_texts, embedding_dimension)
```

```python
def embeddings(text):
  DIM = 300
  zero_vec = np.zeros(DIM)
  features = []
  for tokens in text:
    feat = np.zeros(DIM)
    count = 0 + 1e-5 #for DIV by 0 error
    for token in tokens:
      if token in model:
        feat += model[token]
        count += 1
    if(count !=0):
      features.append(feat/count)
    else:
      features.append(zero_vec)
  return features
```
